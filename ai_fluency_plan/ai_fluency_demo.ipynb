# AI Fluency Demo

This notebook demonstrates a simple LangChain workflow that generates text using the OpenAI LLM. Replace `YOUR_OPENAI_API_KEY` with your actual key.

```python
# Install required packages
!pip install -q langchain openai
```

```python
import os
from langchain import OpenAI, LLMChain

# Set your OpenAI API key
os.environ["OPENAI_API_KEY"] = "YOUR_OPENAI_API_KEY"

# Create an OpenAI LLM instance
llm = OpenAI(temperature=0.7, model_name="gpt-3.5-turbo")

# Define a simple prompt template
prompt_template = """
You are an AI assistant that helps students understand AI fluency concepts.
Explain the importance of data literacy in one paragraph.
"""

# Create LLMChain
chain = LLMChain(llm=llm, prompt=prompt_template)

# Run the chain and print output
response = chain.run()
print(response)
```
